```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

### API global

La API pyplot predeterminada de Matplotlib tiene una interfaz global de estilo MATLAB, como ya hemos visto:

```python
x = np.arange(-10, 11)
```

El fragmento de código crea una figura de 12 x 6 pulgadas y traza dos curvas: ( y = x^2 ) y ( y = -x^2 ). A continuación, se incluye un ejemplo completo para visualizar estas gráficas con matplotlib:

```python
plt.figure(figsize=(12, 6))

plt.plot(x, x**2)
plt.plot(x, -1 * (x**2))
```

Este código crea una figura con dos subgráficos (subplots) en una disposición de 1 fila y 2 columnas.

```python
plt.figure(figsize=(12, 6)) # Esto crea una figura de 12x6 pulgadas
plt.title('Mi bonita trama')# y le asigna el título “Mi bonita trama”.

plt.subplot(1, 2, 1)  # Define el primer subgráfico en una cuadrícula de 1 fila y 2 columnas.
plt.plot(x, x ** 2) # Dibuja la curva ( y = x^2 ).
plt.plot([0, 0, 0], [-10, 0, 100]) #  Dibuja una línea vertical en ( x = 0 ).
plt.legend(['X^2', 'Vertical Line']) # Añade una leyenda para identificar las dos líneas.
plt.xlabel('X') # Etiquetas para los ejes X e Y.
plt.ylabel('X Squared') #  Etiquetas para los ejes X e Y.

plt.subplot(1, 2, 2) # Define el segundo subgráfico en la misma cuadrícula.
plt.plot(x, -1 * (x ** 2)) # Dibuja la curva ( y = -x^2 ).
plt.plot([-10, 0, 10], [-50, -50, -50]) # Dibuja una línea horizontal en ( y = -50 ).
plt.legend(['-X^2', 'Horizontal Line']) # Añade una leyenda para identificar las dos líneas.

plt.xlabel('X') # Etiquetas para los ejes X e Y.
plt.ylabel('X Squared')# Etiquetas para los ejes X e Y.
```

### Interfaz OOP

crea una figura y un conjunto de ejes (subplots) con un tamaño de 12x6 pulgadas.

```python
fig, axes = plt.subplots(figsize=(12, 6))
```

Tu código crea una figura con un solo subgráfico que contiene dos curvas: ( y = x^2 ) y ( y = -x^2 ).
Este código generará una figura con un subgráfico que muestra las dos curvas con sus respectivas configuraciones de color, estilo de línea, marcadores y leyendas.

```python
# Primer gráfico: y = x^2
axes.plot(
    x, (x**2), color='red', linewidth=3,
    marker='o', markersize=8, label='X^2')

# Segundo gráfico: y = -x^2
axes.plot(x, -1 * (x**2), 'b--', label='-X^2')

# Configurar etiquetas y título
axes.set_xlabel('X')
axes.set_ylabel('X Squared')

axes.set_title('Mi bonita trama')

axes.legend()

fig
```

Tu código crea una figura con un subgráfico que muestra cuatro líneas con diferentes estilos de línea:
**sólida, discontinua, guion-punto y punteada**. Cada línea está etiquetada en la leyenda para facilitar la identificación

```python
fig, axes = plt.subplots(figsize=(12, 6))

axes.plot(x, x + 0, linestyle='solid')
axes.plot(x, x + 1, linestyle='dashed')
axes.plot(x, x + 2, linestyle='dashdot')
axes.plot(x, x + 3, linestyle='dotted');

axes.set_title("My Nice Plot")
```

Este código generará una figura con un subgráfico que muestra cuatro líneas con estilos y colores diferentes: **verde sólida, cian discontinua, azul guion-punto y roja punteada.** Cada línea está etiquetada en la leyenda para facilitar la identificación.

```python
fig, axes = plt.subplots(figsize=(12, 6))

axes.plot(x, x + 0, '-og', label="solid green")
axes.plot(x, x + 1, '--c', label="dashed cyan")
axes.plot(x, x + 2, '-.b', label="dashdot blue")
axes.plot(x, x + 3, ':r', label="dotted red")

axes.set_title("My Nice Plot")

axes.legend()
```

Hay muchos tipos de líneas y marcadores.

```python
print('Marcadores: {}'.format([m for m in plt.Line2D.markers]))
```

```python
linestyles = ['_', '--', '-.', ':']

print('Estilos de línea: {}'.format(linestyles))
```

## Otros tipos de parcelas

Cuando llamamos a la función `subplots()` obtenemos una tupla que contiene un elemento `Figure` y un elemento `axes`.

El objeto plot_objects contiene tanto la figura (fig) como los ejes (ax). La línea dibujada conecta los puntos (1,1), (2,2) y (3,3). Al final, plot_objects devuelve la figura y los ejes, que se pueden usar para realizar más modificaciones o visualizaciones.

```python
plot_objects = plt.subplots()

fig, ax = plot_objects

ax.plot([1,2,3], [1,2,3])

plot_objects
```

También podemos definir cuántos elementos queremos dentro de nuestra figura. Para ello podemos configurar los parámetros `nrows` y `ncols`.

Tu código crea una figura con una cuadrícula de 2x2 subgráficos (subplots) y un tamaño de 14x6 pulgadas:

```python
plot_objects = plt.subplots(nrows=2, ncols=2, figsize=(14, 6))

fig, ((ax1, ax2), (ax3, ax4)) = plot_objects

plot_objects
```

Este código generará una figura con cuatro subgráficos, cada uno mostrando una serie de datos aleatorios con diferentes estilos y colores. Cada subgráfico tiene su propio título y leyenda para identificar las líneas:

```python
ax4.plot(np.random.randn(50), c='yellow')
ax1.plot(np.random.randn(50), c='red', linestyle='--')
ax2.plot(np.random.randn(50), c='green', linestyle=':')
ax3.plot(np.random.randn(50), c='blue', marker='o', linewidth=3.0)


fig
```

#### El comando `subplot2grid`

Hay otra forma de crear subtramas utilizando un formato tipo cuadrícula:

Explicación:

**plt.subplot2grid((3, 3), (0, 0),** **colspan=3):** Crea un subgráfico que ocupa toda la primera fila.

**plt.subplot2grid((3, 3), (1, 0), colspan=2):** Crea un subgráfico que ocupa las dos primeras columnas de la segunda fila.

**plt.subplot2grid((3, 3), (1, 2), rowspan=2):** Crea un subgráfico que ocupa la tercera columna de la segunda y tercera fila.

**plt.subplot2grid((3, 3), (2, 0)):**Crea un subgráfico en la primera columna de la tercera fila.

**plt.subplot2grid((3, 3), (2, 1)):** Crea un subgráfico en la segunda columna de la tercera fila.

Cada subgráfico tiene su propio título y leyenda para identificar las líneas.

```python
plt.figure(figsize=(14, 6))

ax1 = plt.subplot2grid((3,3), (0,0), colspan=3)
ax2 = plt.subplot2grid((3,3), (1,0), colspan=2)
ax3 = plt.subplot2grid((3,3), (1,2), rowspan=2)
ax4 = plt.subplot2grid((3,3), (2,0))
ax5 = plt.subplot2grid((3,3), (2,1))
```

### Diagrama de dispersión

**x y y:** Coordenadas aleatorias para los puntos.
**colors:** Colores aleatorios para los puntos.
**area:** Tamaños aleatorios para los puntos, calculados como el área de círculos con radios aleatorios.
Este código generará un gráfico de dispersión donde cada punto tiene un color y tamaño diferente, lo que puede ser útil para visualizar relaciones en datos multidimensionales.

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (20 * np.random.rand(N))**2 # Áreas de los punto
```

**plt.scatter(x, y, c=colors, s=area, alpha=0.5, cmap='Spectral'):** Crea un gráfico de dispersión donde:
x y y son las coordenadas de los puntos.
c=colors asigna colores aleatorios a los puntos.

**s=area** asigna tamaños aleatorios a los puntos.

**alpha=0.5** establece la transparencia de los puntos.

**cmap='Spectral'** utiliza el mapa de colores ‘Spectral’ para los puntos.

**plt.colorbar():** Añade una barra de color para mostrar la escala de colores utilizada.

Este gráfico te permitirá visualizar cómo se distribuyen los puntos con diferentes colores y tamaños.

```python
plt.figure(figsize=(14, 6))

plt.scatter(x, y, c=colors, s=area, alpha=0.5, cmap='Spectral')
plt.colorbar()

plt.show()
```

crea dos graficos de dispersion con los colores que vimos anteriormente:

```python
fig = plt.figure(figsize=(14, 6))

ax1 = fig.add_subplot(1,2,1)
plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Pastel1')
plt.colorbar()

ax2 = fig.add_subplot(1,2,2)
plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Pastel2')
plt.colorbar()

plt.show()
```

### Histogramas

```python
# Datos de ejemplo
values = np.random.randn(1000)
```

**values:** Los datos que se van a representar en el histograma.

**bins=100:** Divide los datos en 100 contenedores (bins).

**alpha=0.8:** Establece la transparencia de las barras del histograma.

**histtype='bar':** Especifica que el histograma se dibuje como barras.

**color='steelblue':** Establece el color de las barras.

**edgecolor='green':** Establece el color del borde de las barras.

```python
# Crear la figura y los ejes
plt.subplots(figsize=(12, 6))

# Crear el histograma
plt.hist(values, bins=100, alpha=0.8,
          histtype='bar', color='steelblue',
          edgecolor='green')

# Establecer los límites del eje X
plt.xlim(xmin=-5, xmax=5)

# Mostrar la figura
plt.show()

```


### KDE (estimación de densidad del núcleo)


Tu código utiliza la función gaussian_kde de la biblioteca scipy para estimar la densidad de probabilidad de los datos en values.

stats.kde.gaussian_kde(values): Esta función calcula una estimación de densidad de Kernel (KDE) para los datos en values. La KDE es una técnica para estimar la función de densidad de probabilidad de una variable aleatoria de manera no paramétrica.

```python
from scipy import stats

density = stats.kde.gaussian_kde(values)
density
```

Este código generará una figura con la curva de densidad de Kernel y un área sombreada debajo de la curva, lo que te permitirá visualizar la distribución de los datos de manera más clara:

```python
plt.subplots(figsize=(12, 6))

# Generar un rango de valores para evaluar la densidad
values2 = np.linspace(min(values)-10, max(values)+10, 100)

# Dibujar la curva de densidad
plt.plot(values2, density(values2), color='#FF7F00')

# Rellenar el área debajo de la curva de densidad
plt.fill_between(values2, 0, density(values2), alpha=0.5, color='#FF7F00')

# Establecer los límites del eje X
plt.xlim(xmin=-5, xmax=5)

# Mostrar la figura
plt.show()
```

### Combinar gráficos

Este código generará una figura con un histograma de los datos en values y una curva de densidad de Kernel superpuesta en color naranja:

Este código crea una figura con un histograma y una curva de densidad de Kernel (KDE) superpuesta.

```python
plt.subplots(figsize=(12, 6))

plt.hist(values, bins=100, alpha=0.8, density=1,
          histtype='bar', color='steelblue',
          edgecolor='green')

plt.plot(values2, density(values2), color='#FF7F00', linewidth=3.0)
plt.xlim(xmin=-5, xmax=5)

plt.show()
```

### Gráficos de barras

np.random.rand(1, 5): Genera una matriz de 1 fila y 5 columnas con valores aleatorios entre 0 y 1.
[0]: Selecciona la primera (y única) fila de la matriz, resultando en un array de 5 elementos.

```python
Y = np.random.rand(1, 5)[0]
Y2 = np.random.rand(1, 5)[0]
```

**np.arange(len(Y)):** Genera un array con los índices de los elementos en Y (0, 1, 2, 3, 4).

**Y:** Los valores de las barras.

**width=barWidth:** Establece el ancho de las barras.

**color='#00b894'**: Establece el color de las barras.

```python
plt.figure(figsize=(12, 4))

barWidth = 0.5
plt.bar(np.arange(len(Y)), Y, width=barWidth, color='#00b894')

plt.show()
```

También se pueden apilar barras y agregar una leyenda al gráfico:

Este código generará un gráfico de barras apiladas con los valores aleatorios en Y y Y2, con barras de diferentes colores y una leyenda para identificar cada conjunto de datos.

```python
plt.figure(figsize=(12, 4))

barWidth = 0.5
plt.bar(np.arange(len(Y)), Y, width=barWidth, color='#00b894', label='Label Y')
plt.bar(np.arange(len(Y2)), Y2, width=barWidth, color='#e17055', bottom=Y, label='Label Y2')

plt.legend()
plt.show()
```

### Diagramas de caja y detección de valores atípicos

Este código genera un histograma de un conjunto de datos que combina valores aleatorios y valores específicos.

**np.random.randn(10):** Genera 10 valores aleatorios de una distribución normal (media 0, desviación estándar 1).

**np.array([10, 15, -10, -15]):** Crea un array con los valores específicos 10, 15, -10 y -15.

**np.concatenate([...]):** Combina los dos arrays en un solo array.

```python
# Generar datos
values = np.concatenate([np.random.randn(10), np.array([10, 15, -10, -15])])
```

```python
# Crear la figura
plt.figure(figsize=(12, 4))

# Crear el histograma
plt.hist(values)
```

Este código genera un diagrama de caja (box plot) para el conjunto de datos values.

```python
# Crear la figura
plt.figure(figsize=(12, 4))

# Crear el diagrama de caja  los datos en values
plt.boxplot(values)
```

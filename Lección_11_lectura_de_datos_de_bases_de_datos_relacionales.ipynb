## Leer datos de una base de datos SQL

Leer datos de bases de datos relacionales SQL es bastante simple y Pandas admite una variedad de métodos para hacerlo.

Comenzaremos con un ejemplo que utiliza SQLite, ya que es un paquete Python integrado y no necesitamos instalar nada adicional.



```python
import sqlite3
```

Para trabajar con una base de datos SQLite desde Python, primero tenemos que conectarnos a ella. Podemos hacerlo mediante la función connect, que devuelve un objeto `Connection`.

Usaremos [esta base de datos de ejemplo](http://www.sqlitetutorial.net/sqlite-sample-database/).

```python
conn = sqlite3.connect("chinook.db")
```

Una vez que tenemos un objeto "Connection", podemos crear un objeto "Cursor". Los cursores nos permiten ejecutar consultas SQL en una base de datos:

```python
cur = conn.cursor()
```

El `Cursor` creado tiene un método `execute`, que recibirá parámetros SQL para ejecutarse en la base de datos.

El código siguiente obtendrá las primeras `5` filas de la tabla `employees`:

```python
cur.execute("SELECT * FROM employees LIMIT 5;")
```

Es posible que hayas notado que no asignamos el resultado de la consulta anterior a una variable. Esto se debe a que necesitamos ejecutar otro comando para obtener los resultados.

Podemos usar el método `fetchall` para obtener todos los resultados de una consulta:

```python
results = cur.fetchall()
```

```python
results
```

Como puede ver, los resultados se devuelven como una lista de tuplas. Cada tupla corresponde a una fila de la base de datos a la que accedimos. Tratar los datos de esta manera es complicado.

Necesitaríamos agregar encabezados de columna manualmente y analizar los datos manualmente. Afortunadamente, la biblioteca de pandas tiene una forma más sencilla, que veremos en la siguiente sección.

```python
df = pd.DataFrame(results)
```

```python
df.head()
```

Antes de continuar, es una buena práctica cerrar los objetos `Connection` y `Cursor` que estén abiertos. Esto evita que la base de datos SQLite se bloquee. Cuando una base de datos SQLite está bloqueada, es posible que no pueda actualizar la base de datos y que se produzcan errores. Podemos cerrar el Cursor y la Connection de esta manera:

```python
cur.close()
conn.close()
```

## Uso del método `read_sql` de pandas

Podemos usar la función `read_sql` de pandas para leer los resultados de una consulta SQL directamente en un `DataFrame` de pandas. El código a continuación ejecutará la misma consulta que acabamos de hacer, pero devolverá un `DataFrame`. Tiene varias ventajas sobre la consulta que hicimos anteriormente:

- No requiere que creemos un objeto `Cursor` ni que llamemos a `fetchall` al final.
- Lee automáticamente los nombres de los encabezados de la tabla.
- Crea un `DataFrame`, por lo que podemos explorar rápidamente los datos.

```python
conn = sqlite3.connect("chinook.db")
```

```python
df = pd.read_sql("SELECT * FROM employees;", conn)
```

```python
df.head()
```

```python
df = pd.read_sql("SELECT * FROM employees;", conn,
                 index_col="EmployeeId",
                 parse_dates=["BirthDate", "HireDate"])
```

```python
df.head()
```

```python
df.info()
```

```python
df["ReportsTo"].isna().sum()
```

```python
df["ReportsTo"].mean()
```

```python
df['ReportsTo'] > 1.75
```

```python
df["City"] = df["City"].astype("category")
```

```python
df.info()
```

## Uso del método `read_sql_query` de pandas

Resulta que el método `read_sql` que vimos anteriormente es solo un contenedor de `read_sql_query` y `read_sql_table`.

Podemos obtener el mismo resultado usando el método `read_sql_query`:

```python
conn = sqlite3.connect("chinook.db")
```

```python
df = pd.read_sql_query("SELECT * FROM employees LIMIT 5;", conn)
```

```python
df.head()
```

```python
df = pd.read_sql_query("SELECT * FROM employees;", conn,
                       index_col="EmployeeId",
                       parse_dates=["BirthDate", "HireDate"])
```

```python
df.head()
```

![green-divider](https://user-images.githubusercontent.com/7065401/52071924-c003ad80-2562-11e9-8297-1c6595f8a7ff.png)


## Uso del método `read_sql_table`

`read_sql_table` es una función útil, pero funciona únicamente con [SQLAlchemy](https://www.sqlalchemy.org/), un kit de herramientas SQL de Python y un mapeador relacional de objetos.

Esta es solo una demostración de su uso, donde leemos toda la tabla `employees`.

```python
from sqlalchemy import create_engine
```

```python
from sqlite3.dbapi2 import connect
engine = create_engine("sqlite:///chinook.db")

connection = engine.connect()
```

```python
df = pd.read_sql_table("employees", con=connection)
```

```python
df.head()
```

```python
df = pd.read_sql_table("employees", con=connection,
                       index_col="EmployeeId",
                       parse_dates=["BirthDate", "HireDate"])
```

```python
df.head()
```

```python
connection.close()
```

## Crear tablas a partir de objetos `DataFrame`

Por último, podemos conservar los objetos `DataFrame` en los que hemos estado trabajando en una base de datos utilizando el método `to_sql` de pandas.

Aunque es fácil de implementar, puede ser un proceso muy lento.

```python
df.head()
```

```python
df.to_sql('employees2', conn, if_exists='replace')
```

```python
pd.read_sql_query("SELECT * FROM employees2;", conn).head()
```

```python
#pd.read_sql_query('DROP TABLE employees2;', conn)
```

## Comportamiento personalizado

El parámetro `if_exists` define cómo comportarse si la tabla ya existe y agrega mucha flexibilidad, permitiéndole decidir si `reemplazar` los datos de la base de datos actual, `agregar` datos nuevos al final o simplemente `fallar` si la base de datos ya existe.

```python
pd.DataFrame().to_sql('employees2',
                      conn,
                      if_exists='replace')
```

```python
pd.read_sql_query("SELECT * FROM employees2;", conn).head()
```

```python
df.to_sql('employees2',
          conn,
          if_exists='replace')
```

```python
pd.read_sql_query("SELECT * FROM employees2;", conn).head()
```

```python
conn.close()
```
